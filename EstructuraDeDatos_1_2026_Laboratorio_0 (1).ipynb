## 8. Tu tarea: completar la implementación

Se debe implementar los siguientes métodos:

- `obtener(posicion)`
- `eliminar(posicion)`
- `mostrar_todos()`
- `calcular_saldo()`

### Recomendaciones
- validar si la posición existe
- decidir cuándo conviene usar `return`
- probar cada método apenas se complete

```python
class ListaMovimientos:
    def __init__(self):
        self.movimientos = []

    def esta_vacia(self):
        return len(self.movimientos) == 0

    def agregar(self, movimiento):
        self.movimientos.append(movimiento)

    def tamano(self):
        return len(self.movimientos)

    def obtener(self, posicion):
        return self.movimientos[posicion]

    def eliminar(self, posicion):
        self.movimientos.pop(posicion)

    def mostrar_todos(self):
        for m in self.movimientos:
            print(m)

    def calcular_saldo(self):
        saldo = 0
        for m in self.movimientos:
            if m["tipo"] == "abono":
                saldo = saldo + m["monto"]
            elif m["tipo"] == "cargo":
                saldo = saldo - m["monto"]

        return saldo
```
