## 3. Conceptos iniciales

### Definiciones básicas
- **Clase:** el molde o definición general.
- **Objeto:** una instancia concreta creada a partir de una clase.
- **Atributo:** una característica o dato del objeto.
- **Método:** una acción o comportamiento del objeto.
- **Interfaz (idea inicial):** el conjunto de acciones con las que se puede interactuar con el objeto.

## 5. Completa la clase

Ahora agrega métodos simples a la clase.

### Tareas
- Crear un método `mostrar_info()` que imprima la información principal de la sala.
- Crear un método `reservar()` que deje la sala como no disponible si estaba libre.
- Crear un método `liberar()` que deje la sala disponible nuevamente.

### Pista
Recuerda que dentro de los métodos puedes acceder a los atributos usando `self`.


```python
class Sala:
    def __init__(self, nombre, capacidad, ubicacion, disponible=True):
        self.nombre = nombre
        self.capacidad = capacidad
        self.ubicacion = ubicacion
        self.disponible = disponible

    # Completa este método
    def mostrar_info(self):
        print("Nombre:", self.nombre)
        print("Capacidad:", self.capacidad)
        print("Ubicación:", self.ubicacion)
        print("¿Está disponible?:", self.disponible)

    # Completa este método
    def reservar(self):
        self.disponible = False
        print("¡La sala", self.nombre, "ha sido reservada!")

    # Completa este método
    def liberar(self):
        self.disponible = True
        print("La sala", self.nombre, "ahora está libre de nuevo.")
```
